# Explanation
This section contains a more detailed explanation of how image and asnwer is scored

```python
from AnswerScoring import *
from HandwritingRecognition import seperate_each_line
```

### 1) Image Preprocessing
The Handwriting OCR model would only work on single lines of text. However, a student's anwer will contain multiple lines so we used OpenCV to split each line of text into seperate images. We also made the image more clear to increase the accuracy of the model.

```python
line_images = seperate_each_line(answer_img_2)
```

```python
fig, axs = plt.subplots(len(line_images))
fig.suptitle('PreProcessed Images')
for i in range(len(line_images)):
    axs[i].imshow(line_images[i])
plt.show()
```

### 2) Image OCR
Pass the images to the OCR model and convert them to text

```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel

answer_text = ""
processor = TrOCRProcessor.from_pretrained('microsoft/trocr-base-handwritten')
model = VisionEncoderDecoderModel.from_pretrained('microsoft/trocr-base-handwritten')
for line in line_images:
    try:
        pixel_values = processor(images=line, return_tensors="pt").pixel_values
        generated_ids = model.generate(pixel_values)
        generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
        answer_text += generated_text + " \n"
    except:
        pass

print("-----------------------------------")
print(f"The answer text is:\n{answer_text}")
print("-----------------------------------")
```

### 3) Named Entity & Verb Recognition
We run both the student's answer and the sample answer through a named entity detection (NER) model. This helps us pick up names, places, dates, etc. so that we can make a proper comparison while scoring later.

```python
answer_text_tokens = do_tokenisation(answer_text)
verbs = get_verbs(answer_text)
answer_text_tokens['VERBS'] = verbs
marking_scheme_tokens = [do_tokenisation(marking_scheme_element) for marking_scheme_element in marking_scheme]

print(f"Student Answer NER = {answer_text_tokens}")
print(f"Marking Scheme NER = {marking_scheme_tokens}")
```

### 4) Compare Answers (with Synonym Checking)

```python
score = 0
explaination = ""
for point in marking_scheme:
    result, error = compare_answer(point, answer_text_tokens)
    print("Result = ", result, "; Error = ", error)
    if result:
        score += 1
    else:
        explaination += error
```

```python
print(f"score = {score} (which is out of {len(marking_scheme)})\nexplaination = {explain}")
```
